```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(0)
df=pd.DataFrame({
    'Category': ['A','B','C','D','E'],
    'VALUES': [10,15,20,25,30],
    'SALES': np.random.randint(100,500,5),
    'Profit': np.random.randint(50,200,5)
})
print(df.head())
```

```python
#1. Area Plot
plt.figure()
plt.fill_between(df['Category'],df['VALUES'],color='green',alpha = 1)
plt.title("Area Plot")
plt.xlabel("Category")
plt.ylabel("Values")
plt.show
```

```python
#2. Pie CHart
plt.figure()
plt.pie(df['VALUES'],labels=df['Category'],autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

```python
#3. Donut chart
plt.figure()
plt.pie(df['VALUES'],labels=df['Category'],autopct='%1.1f%%',)
centre_circle=plt.Circle((0,0),0.70,fc='white')
fig=plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Donut Chart")
plt.show()
```

```python
#4. Boolean (Outliers detection)
plt.figure()
plt.boxplot(x=df['VALUES'])
plt.title("Boxplot of Sales")
plt.show()
```

```python
#5 Heatmap (correlation)
import seaborn as sns

plt.figure()
corr= df[['SALES' ,'Profit']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#6  Bubble plot
import matplotlib.pyplot as plt
plt.scatter(df['SALES'], df['Profit'], s=df['VALUES']*10) #size of bubbles
plt.title("Bubble Plot")
plt.xlabel("SALES")
plt.ylabel("Profit")
plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x='SALES', y='Profit',
                size='VALUES',     #bubble size
                hue='VALUES',      #color based on values
                data=df,
                sizes=(50,300),   #min max size
                palette='viridis')
plt.title("Bubble Plot using Seaborn")
plt.show()
```

```python
## 2. Second Dataset
import numpy as np
import pandas as pd
np.random.seed(7)
n=100
df = pd.DataFrame({
    'Customer_ID': range(1, n+1),
    'Age':np.random.randint(20,60,n),
    'Income':np.random.randint(20000,100000,n),
    'Loan_Amount':np.random.randint(5000,50000,n),
    'Credit_Score':np.random.randint(300,900,n)
})

print(df.head())
```
